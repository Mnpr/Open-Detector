**Mean Average Precision** `mAP`
- Average precision is calculated for each classes
- The mean of average precisions across all classes `mAP`to evaluate accuracy

**Terms**

- `Intersection Over Union(IoU)` is calculated by measuring overlaping area between ground truth and predicted boundary boxes.
  - IoU : `Area of overlap` / `Area of Union`
- `Threshold` value to identify bbox-prediction as `True` or `False`
- True Positive(`TP`) : Number of correct b-box predictions [ IoU $\ge$ Threshold ]
- False Positive(`FP`) : Number of wrong b-box predictions [IoU $\lt$ Threshold ]
- False Negative(`FN`) : Nuber of un-detected ground truth b-boxes
- True Negative(`TN`) : Number of b-box that are incorrectly predicted

**Precision**

- % of correct positive predictions in all detected b-boxes. i.e. $Precision = \frac{TP}{TP + FP} $

**Recall**

- % of  `TP` prediction among all relevant ground truth b-boxes. i.e. $Recall = \frac{TP}{TP + FN}$

**Precision-Recall Curve**

- It represents the performance of the model by plotting the curve (`precision` vs `recall`). 
- Accumulate detected b-box with confidences in descending order
- Calculate precision and recall for every detected b-boxes

**Average Precision**

The weighted mean of precision at each threshold 

$$
\text{AP} = \sum_n (R_n - R_{n-1}) P_n
$$


```python
fig = plt.figure()

plt.plot([0, 0.2, 0.2, 0.4, 0.6, 0.6, 0.8, 1, 1, 1, 1], 
         [1, 1, 0.5, 0.67, 0.75, 0.6, 0.67, 0.71, 0.62, 0.56, 0.5],'-o', linewidth=3.0, color='blue', markersize=10.0)

plt.xlabel('Recall', fontsize=20)
plt.ylabel('Precision', fontsize=20)
plt.tick_params(labelsize=18)
plt.title('Recall and Precision curve', fontsize=22)
plt.show()
```
